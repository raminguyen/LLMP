## Display images for each task to verify if it correctly label it

```python
import json
import os
import pandas as pd
from PIL import Image
import matplotlib.pyplot as plt

# Load the combined JSON dataset
json_file = "output/json/combined_dataset.json"
image_folder = "output/images"

with open(json_file, 'r') as f:
    data = json.load(f)

# Set the specific task for 'area'
task = "Please estimate the area covered by the circle. The whole image is 100x100 with an area of 10000. No explanation."

# Convert the JSON data to a DataFrame
df = pd.DataFrame(data)

# Function to display images for a particular task with a tighter grid layout
def display_images_for_task(df, task, grid_rows=5, grid_cols=5):
    # Filter DataFrame for the specific task
    task_df = df[df['question'] == task]

    # Number of images to display (max is rows * cols)
    max_images = grid_rows * grid_cols
    num_images = min(len(task_df), max_images)

    # Create a plot with subplots for the grid
    fig, axes = plt.subplots(grid_rows, grid_cols, figsize=(10, 10))

    # Flatten axes for easier iteration
    axes = axes.flatten()

    # Loop through images and their corresponding values
    for idx, (i, row) in enumerate(task_df.iterrows()):
        if idx >= num_images:
            break

        image_path = os.path.join(image_folder, row['image'])
        img = Image.open(image_path)

        # Display the image in the grid
        axes[idx].imshow(img, cmap='gray')
        axes[idx].set_title(f"Value: {row['value']}", fontsize=6)
        axes[idx].axis('off')  # Hide axes for cleaner display

    # Turn off any extra axes (if less than grid size)
    for i in range(num_images, len(axes)):
        axes[i].axis('off')

    # Adjust layout to reduce spacing between subplots
    plt.subplots_adjust(wspace=-0.5, hspace=-0.5)  # Reduce the space between images
    plt.tight_layout(pad=0)  # Remove padding around the figure
    plt.show()

# Display images only for the 'area' task
display_images_for_task(df, task, grid_rows=5, grid_cols=5)
```

```python
# Loop through each unique task in the dataset and plot the value distribution
for task in df['question'].unique():
    print(f"Plotting value distribution for task: {task}")
    
    # Filter DataFrame for the specific task
    task_df = df[df['question'] == task]

    # Extract the values (labels) for the task
    values = task_df['value']

    # Plot the distribution of values
    plt.figure(figsize=(10, 6))
    plt.hist(values, bins=20, color='green', edgecolor='black')  # Adjust bins as necessary
    plt.title(f"Value Distribution for Task: {task}", fontsize=14)
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```
